```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
# Columns Required for content based movie analysis
#1. Genere.
#2. Keywords.
#3. Title.
#4. Overview.
#5. Cast.
#6. Crew.
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#Removing Missing Data
movies.dropna(inplace = True)
```

```python
#Checking is missing data is removed or not
movies.isnull().sum()
```

```python
#Checking for duplicate data.
movies.duplicated().sum()
```

```python
#Shaping genres column into simple list
movies.iloc[0].genres
```

```python
#Shaping genres column into simple list
def convert(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#converting string of list into list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Shaping genres column into simple list
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#Converting Genres into list using convert function
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#Converting Keywords into list using convert function
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#Creating a function which will extract 3 cast name from cast column.
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#Creating function which will fetching Director name from crew column
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if (i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
            
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#Converting Overview column into list so that we can concat with other list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Doing transformation on columns which will remove the spaces between between entities.
#i.e name = 'Sam Worthington' we need to remove spaces between them
#Otherwise our model will be confused as to which name it should recommend
#there can be multiple with sam as it takes the first enitiy i.e sam from Sam Worthington
```

```python
#Removing spaces in genres,keywords,cast and crew columns in our datasets
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#Creating a column name Tag where we will concat all the columns into it
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
#Creating a new data frame with movie_id, title and tags column
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#Converting tags into string as it is in list
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#Converting tags column into lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
#Using NLP Stemming algoritm to remove words such as  'alcohol','alcoholic','alcoholism' and keep only one word
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#Example of Stemming with love words
```

```python
ps.stem('love')
```

```python
ps.stem('loved')
```

```python
#Converting Text into vector using Bag Of Words NLP Algorithm

#Removing stop words from tags such as is,are,from etc using scikit learn.

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#Converting cv object into numpy array as it returns sparse matrix
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
#Checking all the frequent words from our corpus
cv.get_feature_names()
```

```python
#We will use cosine similarity to calculate angle between vectors 
#We cannot use euclidean distance as it fails to give appropriate 
#measures when dealing with large dimensional data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#Performing decending sorting based on movies index and finding the distances similar
#to the the particular movie index and getting 5 movies similar to the search movie
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
#Fteching the index and getting the similarity of the movies
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:5]
    for i in movies_list:
        #Getting movie recommendation and printing the title
        print(new_df.iloc[i[0]].title)
     
```

```python
recommend('Avatar')
```
